 # Accuracy vs Prediction Set Size(model fix and varying methods)

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Load the CSV file
file_path = 'comprehensive_predictions_summary_4.csv'  # replace with your file path
df = pd.read_csv(file_path)

# Extract unique models and methods
models = df['Model'].unique()
methods = df['Method'].unique()

# Exclude methods 'IID' and 'biased(CE)'
methods = [method for method in methods if method not in ['IID', 'biased(CE)']]

# Define symbols for different methods
symbols = ['circle', 'square', 'diamond', 'cross', 'x', 'triangle-up', 'triangle-down']

# Loop through each model and create a subplot figure for each
for model in models:
    # Filter data for the current model
    model_data = df[df['Model'] == model]

    # Create a figure with subplots (one subplot for each method)
    fig = make_subplots(
        rows=1, cols=len(methods),  # Subplots for the number of methods
        subplot_titles=methods  # Titles for each subplot (methods)
    )

    # Loop through each method and add scatter plots in each subplot
    for i, method in enumerate(methods):
        method_data = model_data[model_data['Method'] == method]
        
        fig.add_trace(
            go.Scatter(
                x=method_data['Accuracy_mean'],  # Accuracy on the x-axis
                y=method_data['Prediction_Size_90_mean'],  # Prediction Set Size on the y-axis
                mode='markers',  # Keep scatter plot
                name=method,
                marker=dict(symbol=symbols[i], size=10),  # Different symbols for each method
                hovertemplate=(  # Customize hover information
                    'Method: %{text}<br>' +
                    'LD1: %{customdata[0]}<br>' +
                    'LD2: %{customdata[1]}<br>' +
                    'Alpha Bias: %{customdata[2]}<br>' +
                    'Accuracy: %{x}<br>' +
                    'Prediction Set Size: %{y}<extra></extra>'
                ),
                text=method_data['Method'],  # Hover display of the method name
                customdata=method_data[['LD1', 'LD2', 'Alpha Bias']]  # Additional hover data
            ),
            row=1, col=i+1  # Position of the subplot (row=1, col=i+1)
        )

    # Update layout for the figure
    fig.update_layout(
        title=f"Accuracy vs Prediction Set Size for Model: {model}",
        height=600,
        width=1400,  # Adjust width for subplots
        font=dict(size=14),
        showlegend=False,  # No legend needed (one plot per method)
        yaxis_title="Prediction Set Size"  # Common y-axis title for all subplots
    )

    # Manually adjust the x-axis title for each subplot and position the title
    for i in range(1, len(methods) + 1):
        fig.update_xaxes(title_text="Accuracy", row=1, col=i, title_standoff=10)

    # Set the same range for all x and y axes
    fig.update_yaxes(range=[2.0, 5])  # Adjust x-axis range to [2.1, 5]
    fig.update_xaxes(range=[0.6, 0.86])
    
    # Show the figure for the current model
    fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load the CSV file
file_path = 'comprehensive_predictions_summary_citeseer.csv'
df = pd.read_csv(file_path)

# Convert 'Accuracy_mean' to numeric
df['Accuracy_mean'] = pd.to_numeric(df['Accuracy_mean'], errors='coerce')

# Set the specific accuracy value for the horizontal line
specific_accuracy = 0.561

# Create a figure with higher resolution
fig = go.Figure()

# Define color map for models
color_map = {
    'APPNP': '#7B68EE',
    'DAGNN': '#FF6347',
    'GAT': '#3CB371',
    'GCN': '#9370DB',
    'GraphSAGE': '#FFA500'
}

# Define the order of methods
method_order = ['IID', 'biased(CE)', 'SRGNN', 'cmd', 'MMD', 'CJS', 'CKL', 'jsd', 'kld']

# Loop through each unique model
models = ['APPNP', 'DAGNN', 'GAT', 'GCN', 'GraphSAGE']

for model in models:
    model_data = df[df['Model'] == model]

    # Add bars with improved text formatting
    fig.add_trace(go.Bar(
        x=method_order,
        y=[model_data[model_data['Method'] == method]['Accuracy_mean'].values[0] if not model_data[model_data['Method'] == method].empty else None for method in method_order],
        name=model,
        marker_color=color_map[model],
        text=[f"{model_data[model_data['Method'] == method]['Accuracy_mean'].values[0]:.3f}" if not model_data[model_data['Method'] == method].empty else '' for method in method_order],
        textposition='outside',
        textfont=dict(
            size=11,  # Increased text size
            color='black',  # Explicit text color
            family='Arial'  # Clear font family
        ),
        hoverinfo='text',
        hovertext=[f"{model}: {model_data[model_data['Method'] == method]['Accuracy_mean'].values[0]:.3f}" if not model_data[model_data['Method'] == method].empty else '' for method in method_order],
    ))

# Add a dashed line for the overall minimum value
fig.add_shape(
    type="line",
    x0=-0.5,  # Start of the line (leftmost point)
    x1=len(method_order) - 0.5,  # End of the line (rightmost point)
    y0=overall_min,
    y1=overall_min,
    line=dict(
        color="grey",
        width=2,
        dash="dash"  # Dashed line
    )
)

# Update layout with improved visibility
fig.update_layout(
    title={
        'text': "Accuracy vs Methods (Grouped by Model)",
        'y': 0.95,
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top',
        'font': dict(size=16)  # Larger title font
    },
    xaxis_title=dict(
        text="Method",
        font=dict(size=14)  # Larger axis title
    ),
    yaxis_title=dict(
        text="Accuracy",
        font=dict(size=14)  # Larger axis title
    ),
    height=600,  # Increased height
    width=1200,  # Increased width
    barmode='group',
    font=dict(size=12),
    showlegend=True,
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1,
        font=dict(size=12)  # Legend font size
    ),
    plot_bgcolor='rgba(242,242,242,1)',
    yaxis=dict(
        tickformat='.2f', 
        range=[0, 1],
        dtick=0.2,
        gridcolor='white',
        zerolinecolor='white',
        tickfont=dict(size=12)  # Axis tick labels font size
    ),
    xaxis=dict(
        tickangle=45,  # Angled labels for better readability
        gridcolor='white',
        tickfont=dict(size=12)  # Axis tick labels font size
    ),
    bargap=0.15,
    bargroupgap=0.05
)

# Improve resolution for export
fig.update_layout(
    template="plotly_white",
    paper_bgcolor='white',
    plot_bgcolor='rgba(242,242,242,1)'
)

# Show the figure
fig.show()

# Optional: Save as high-resolution image
fig.write_image("high_res_plot_with_specific_accuracy.png", scale=10)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the dataset
file_path = 'comprehensive_predictions_summary_4.csv'  # Update with your file path
df = pd.read_csv(file_path)

# Display basic info and check for missing values
print("Dataset Info:\n", df.info())
print("\nSummary Statistics:\n", df.describe())

# Drop rows with missing values
df = df.dropna()

# ===========================================================
# 1. Correlation Analysis
# ===========================================================
# Compute correlation matrix
# Identify non-numeric columns
non_numeric_cols = df.select_dtypes(include=['object']).columns
print("Non-Numeric Columns:\n", non_numeric_cols)

# Drop non-numeric columns for correlation analysis
df_numeric = df.select_dtypes(include=['number'])

# Compute correlation matrix for numeric data only
correlation_matrix = df_numeric.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()


# ===========================================================
# 2. Scatterplots: Lambda1, Lambda2, and Accuracy
# ===========================================================
sns.scatterplot(data=df, x='LD1', y='LD2', size='Accuracy_mean', hue='Accuracy_mean', palette="viridis")
plt.title("Lambda1 vs Lambda2 with Accuracy")
plt.show()

# ===========================================================
# 3. Pairplot for Relationships
# ===========================================================
# Select numerical columns for pairplot
numerical_cols = ['LD1', 'LD2', 'Accuracy_mean', 'Coverage', 'Prediction_Size']

sns.pairplot(df[numerical_cols], diag_kind="kde", hue="lambda1", palette="coolwarm")
plt.show()

# ===========================================================
# 4. 3D Visualization: Lambda1, Lambda2, and Accuracy
# ===========================================================
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter 3D
scatter = ax.scatter(df['LD1'], df['LD2'], df['Accuracy_mean'], c=df['Accuracy_mean'], cmap='viridis', s=50)
ax.set_xlabel('LD1')
ax.set_ylabel('Lambda2')
ax.set_zlabel('Accuracy_mean')
plt.colorbar(scatter, ax=ax, shrink=0.5, aspect=5)
plt.title("3D Scatter: LD1, Lambda2, Accuracy")
plt.show()

# ===========================================================
# 5. Boxplot Analysis: Accuracy by Lambda1 and Lambda2
# ===========================================================
sns.boxplot(data=df, x='lambda1', y='Accuracy_mean', hue='LD2', palette="Set3")
plt.title("Accuracy Distribution by LD1 and Lambda2")
plt.show()

# ===========================================================
# 6. Metric Insights: Top Configurations
# ===========================================================
# Find top 10 configurations based on Accuracy_mean
top_results = df.nlargest(10, 'Accuracy_mean')
print("\nTop Configurations with Highest Accuracy:")
print(top_results[['LD1', 'LD2', 'Accuracy_mean', 'Coverage', 'Prediction_Size']])

# ===========================================================
# 7. Coverage and Prediction Size Trends
# ===========================================================
# Line plot for Coverage and Prediction_Size against Accuracy
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Accuracy_mean', y='Coverage', label='Coverage', marker='o')
sns.lineplot(data=df, x='Accuracy_mean', y='Prediction_Size', label='Prediction Size', marker='s')
plt.title("Coverage and Prediction Size vs Accuracy")
plt.xlabel("Accuracy_mean")
plt.ylabel("Metrics")
plt.legend()
plt.show()

# ===========================================================
# 8. Jointplot: Lambda Interaction with Metrics
# ===========================================================
sns.jointplot(data=df, x='LD1', y='Coverage', hue='Accuracy_mean', kind='scatter', palette="coolwarm")
plt.title("Lambda1 vs Coverage Colored by Accuracy")
plt.show()

sns.jointplot(data=df, x='LD2', y='Prediction_Size', hue='Accuracy_mean', kind='scatter', palette="viridis")
plt.title("Lambda2 vs Prediction Size Colored by Accuracy")
plt.show()
```
